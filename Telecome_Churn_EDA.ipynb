## replacing blanks with 0 and coverting it to float 

```python
df['TotalCharges'] = df['TotalCharges'].replace(' ','0')
```

```python
df['TotalCharges'] = df['TotalCharges'].astype(float)
```

```python
df.info()
```

## percentage of male and female customers

```python
df.value_counts('gender').plot(kind = 'pie', autopct ='%.2f%%', title = 'Percentage of male and female customers')
```

## convert 0 and 1 values to yes/no 

```python
def customer_type(a):
    if a == 0:
        return 'no'
    else:
        return 'yes'

df['SeniorCitizen'] = df['SeniorCitizen'].apply(customer_type)
```

```python
df.head()
```

```python
df.value_counts('SeniorCitizen').plot(kind = 'pie', autopct = '%.2f%%', title = "Proportion of Senior vs. Non-Senior Citizens")
```

```python
df.value_counts('Churn').plot(kind = 'pie', autopct ='%.2f%%',title = 'Propotion of Churn out vs non-Churn out customers')
```

```python
sns.countplot(x=df['Churn'])
plt.title('Count of customers by Churn out')
```

```python
sns.countplot(hue = 'gender', x = 'Churn', data = df)
plt.title('Customer Churn by Gender')
plt.show()
```

```python
sns.countplot(hue = 'Churn', x = 'SeniorCitizen', data = df)
plt.title('Customer Churn by Senior Citizen')
plt.show()
```

```python
churn_by_senior = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Plot as a stacked bar chart
churn_by_senior.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'])

# Add title and labels
plt.title('Churn by Senior Citizen Status')
plt.xlabel('Senior Citizen (0 = No, 1 = Yes)')
plt.ylabel('Number of Customers')
plt.legend(title='Churn')
plt.show()
```

```python


# Columns to visualize
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up figure size and layout (3 rows × 3 columns)
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()  # flatten the 2D array of axes

# Loop through columns and axes to plot each
for i, col in enumerate(cols):
    sns.countplot(x=col, data=df, ax=axes[i], palette='Set2')
    axes[i].set_title(f'{col} Count', fontsize=12, fontweight='bold')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# Remove any unused subplot (if number of plots < grid size)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust spacing between plots
plt.tight_layout()
plt.show()
```

```python


# Columns to visualize
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set Seaborn style
sns.set(style="white")

# Create subplots (3 rows × 3 columns)
fig, axes = plt.subplots(3, 3, figsize=(16, 12))
axes = axes.flatten()

# Loop through columns
for i, col in enumerate(cols):
    sns.countplot(
        x=col, hue='Churn', data=df,
        ax=axes[i], palette='Set2'   # <-- darker palette
    )
    axes[i].set_title(f'{col} vs Churn', fontsize=12, fontweight='bold')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')
    axes[i].legend(title='Churn', loc='upper right', fontsize=9)

# Remove unused subplot if any
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()

```
